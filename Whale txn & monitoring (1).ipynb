```python
import time

wallets = [
    '0x5a52e96bacdabb82fd05763e25335261b270efcb',  # whale 1
    '0xa9ac43f5b5e38155a288d1a01d2cbc4478e14573',  # Whale wallet 2
    '0xfe908731af54c56dfd63948f18ce5878a3b8a1ad'   # Whale wallet 3
]

def get_large_eth_transfers(address, threshold=1000):
    url = f'https://api.etherscan.io/api?module=account&action=txlist&address={address}&startblock=0&endblock=99999999&sort=desc&apikey={ETHERSCAN_API_KEY}'
    response = requests.get(url)
    data = response.json()

    if data['status'] != '1':
        print(f"❌ {address} - {data['message']}")
        return []

    large_txs = []
    for tx in data['result']:
        eth_value = int(tx['value']) / 10**18
        if eth_value >= threshold:
            large_txs.append({
                'from': tx['from'],
                'to': tx['to'],
                'value': eth_value,
                'timestamp': tx['timeStamp']
            })

    return large_txs

# Loop through wallets
for wallet in wallets:
    print(f"\n🔍 Checking wallet: {wallet}")
    txs = get_large_eth_transfers(wallet)
    if txs:
        for tx in txs:
            print(f"🟢 {tx['from']} → {tx['to']} | {tx['value']}")
    else:
        print("No large transfers found.")
    time.sleep(1)  # Respect Etherscan's rate limit
```
